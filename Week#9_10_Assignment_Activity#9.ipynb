```python
import urllib.request
import urllib.parse
from urllib.error import HTTPError,URLError
import json
import requests
from bs4 import BeautifulSoup
import re
from OpenSSL import SSL
```

```python
#Chceking SSL certifications using OpenSSL:
ctx = SSL.Context(SSL.SSLv23_METHOD)
ctx.check_hostname = False
ctx.verify_mode = SSL.VERIFY_NONE
```

```python
url = "http://www.gutenberg.org/browse/scores/top"
response = requests.get(url)
```

```python
def checkRequestStatus(response):
    return response.status_code
```

```python
content = response.content.decode(response.encoding)
soup = BeautifulSoup(content,'html.parser')
```

```python
for h2 in soup.find_all('h2'):
    if h2.text == "Top 100 EBooks yesterday":
        print(h2,h2.find_all('ol'))
        
```

```python
list_link = []
for links in soup.find_all('a'):
        list_link.append(links.get('href'))
```

List of book starts from 19 to 119 in the list

```python
book_num = []
for links in list_link[19:119]:
    regex_int = "[0-9]+"
    book_no = int(re.findall('[0-9]+',links)[0])
    book_num.append(book_no)
```

```python
soup_lines = soup.text.split('\n')
#soup_lines
pos_start = soup_lines.index("Top 100 EBooks yesterday")
```

```python
top_100_book_names = soup_lines[pos_start+2:pos_start+2+100]
```

```python
top_100_book_names
```

```python
final_list = []
for i in range(100):
    start,end = re.match('^[a-zA-Z ]*',top_100_book_names[i]).span()
    final_list.append(top_100_book_names[i][start:end])
    
```
